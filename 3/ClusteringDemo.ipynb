# Imports

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.datasets.samples_generator import make_blobs
```

# Create Blobs

```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.4,
                            random_state=0)

X = StandardScaler().fit_transform(X)

colors = [[0.61960787, 0.00392157, 0.25882354, 1.],
[0.99346405, 0.74771243, 0.43529413, 1.],
[0.74771243, 0.89803922, 0.627451, 1.]]
```

# Plot Function

```python
def plot(X, labels):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)

    for p, lbl in zip(X, labels):
        if lbl == -1:
            c = 'k'
        else:
            c = colors[lbl]
            
        ax.scatter(p[0],p[1], color=c, s=100, edgecolors='k')
```

---
# BLOBS EXAMPLE
----

# K Means

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
labels = km.fit_predict(X)


plot(X, labels)
```

# Mean Shift

```python
from sklearn.cluster import MeanShift
ms = MeanShift()
labels = ms.fit_predict(X)
plot(X, labels)
```

# DBSCAN

```python
db = DBSCAN(eps=0.3, min_samples=10)
labels = db.fit_predict(X)

plot(X, labels)
```

# Ground Truth

```python
from sklearn.datasets import load_iris
from sklearn.decomposition.pca import PCA

X = load_iris()['data']
pca = PCA(2)
rX = pca.fit_transform(X)
plot(rX, load_iris()['target'])
```

# K Means

```python
km = KMeans(n_clusters=3)
labels = km.fit_predict(X)

plot(rX, labels)
```

# Mean Shift

```python
from sklearn.cluster import mean_shift
ms = MeanShift()
labels = ms.fit_predict(X)

plot(rX, labels)
```

# DBSCAN

```python
db = DBSCAN(eps=0.6, min_samples=5)
labels = db.fit_predict(X)

plot(rX, labels)
```
