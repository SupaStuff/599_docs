```python
## Synth Data

x = np.random.random(1000)
```

```python
true_b1 = 3
true_b0 = 1
noise_std = 0.1

y = true_b1 * x + true_b0 + np.random.randn(1000)*noise_std
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x,y)
```

```python
X = np.vstack([x,np.ones(len(x))])
```
